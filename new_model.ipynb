```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = stopwords.words('english')

# find the words other than stopwords 
def Find_words(title):
    lst=[]
    for i in word_tokenize(title):
        if i not in stop_words:
            lst.append(i)
    return lst

try_df['new_title']=try_df['title'].apply(Find_words)
```

```python
try_df['relation']=-1
try_df
```

```python
# find no. of words that are common in the titlee and text
def Find_relation(idx):
    cnt=0
    text = try_df.iloc[idx]['text']
    for i in try_df.iloc[idx]['new_title']:
        if i in text:
#             print(i)
            cnt+=1
    try_df.at[idx,'relation']=cnt
#     print(try_df.iloc[idx]['new_title'])


for i in try_df['new_title'].index:
    Find_relation(i)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
# nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

lemmatizer=WordNetLemmatizer()

def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:         
        return None
```

```python
def lemm(text):
    sent=""
    for i in text.split():
        lst=[i]
        pos1=nltk.pos_tag(lst)
        
#         print(type(pos1),pos1) => <class 'list'> [('takes', 'VBZ')]
        
        tag= pos_tagger(pos1[0][1])
#         print(i)
        if tag==None or i=="warning":
            sent=sent + " " + i
        else:
            sent=sent + " " + lemmatizer.lemmatize(i,pos=tag)
#             print(lemmatizer.lemmatize(i,pos=tag))
    return sent    
```

```python
try_df['input']=try_df['input'].apply(lemm)
```
